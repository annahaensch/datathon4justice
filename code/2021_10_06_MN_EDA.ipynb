```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/MN_sentencing/allmn.csv", dtype={45:str, 46:str})
```

```python
for y in df["sentyear"].unique():
    df_y = df[df["sentyear"] == y]
    df_y.reset_index(inplace = True)
    df_y.to_csv("../data/MN_sentencing/{}.csv".format(y))
```
